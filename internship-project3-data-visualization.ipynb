```python
#Load Libraries for Data Analysis
import pandas as pd
import os
os.listdir('/kaggle/input/datasets/vincenthope/stockprice')
```

```python
data = pd.read_csv("/kaggle/input/datasets/vincenthope/stockprice/Stock_Prices.csv")

data.head(5)
```

```python
import matplotlib.pyplot as plt

corr = data[['open','high','low','close','volume']].corr()

plt.bar(corr.index, corr['close'])
plt.title("Correlation with Close Price")
plt.show()
```

```python
aapl = data[data['symbol'] == 'AAPL']

plt.figure()
plt.plot(aapl['date'], aapl['close'], label="Close Price")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("AAPL Close Price Over Time")
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("line_chart.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

data['close'].plot(kind='line')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='open', y='close', data=data)

plt.title("Open vs Close (Seaborn)")
plt.savefig("scatter_seaborn.png")
plt.show()
```

```python
plt.savefig("plot.png", dpi=300)   # high quality
plt.savefig("plot.pdf")            # vector format (best for reports)
```

```python
plt.figure()

plt.plot(aapl['date'], aapl['open'], label='Open')
plt.plot(aapl['date'], aapl['close'], label='Close')

plt.legend()
plt.title("Open vs Close Prices")
plt.xticks(rotation=45)

plt.savefig("multi_line.png")
plt.show()
```

```python
#Load Libraries for Data Analysis
import pandas as pd
import os
os.listdir('/kaggle/input/datasets/vincenthope/stock-prices/')
```

```python
data = pd.read_csv("/kaggle/input/datasets/vincenthope/stock-prices/Stock_Prices.csv")

data.head(5)
```

```python
# Check for missing values (True/False)
data.isnull()
```

```python
# Counting missing values per column
data.isnull().sum()
```

```python
# Total missing values
data.isnull().sum().sum()
```

```python
# Check How Many Records There Are 
data.shape 
```

```python
# Drop Rows With Missing Values Since Not Much i.e 27
data_clean = data.dropna()

# Check to see output 
data_clean.isnull().sum()

```

```python
# Check How Many Records Remain
data_clean.shape 
```

```python
# Check for Duplicates - Shows True for Duplicate Rows
data_clean.duplicated()
```

```python
# Check for No of Duplicates 
data_clean.duplicated().sum()
```

```python
# Display first 10 Records

data_clean.head(10)
```

**Standardizing Inconsistent Data Formats is about making your data uniform and clean****.

```python
# Standardize Date Records
data_clean.loc[:, 'date'] = pd.to_datetime(data_clean['date'])
```

```python
# Display first 10 Records

data_clean.head(10)
```

```python
# To Standardize Text Data
# strip() → removes spaces
# title() → capitalizes properly

data_clean.loc[:, 'symbol'] = data_clean['symbol'].str.strip().str.title()
```

```python
# Display first 10 Records

data_clean.head(10)
```

```python
# Convert numeric column
data_clean.loc[:, 'volume'] = pd.to_numeric(data_clean['volume'], errors='coerce')
```

```python
# Display first 10 Records

data_clean.head(10)
```

```python
# Export Clean & Processed Data
# index = false : does not save row nos as columns 
data_clean.to_csv("Project1.csv", index=False)
print("Data saved successfully!")
```

```python
#Load Libraries for Data Analysis
import pandas as pd
import os
os.listdir('/kaggle/input/datasets/vincenthope/stockprice')
```

```python
data = pd.read_csv("/kaggle/input/datasets/vincenthope/stockprice/Stock_Prices.csv")

data.head(5)
```

```python
# Calculate Summaary Statisics 
data.describe()
```

```python
# Histogram shows distribution of ONE column
import matplotlib.pyplot as plt
plt.hist(data['open'], bins=10, alpha=0.5, label='high', color='blue', edgecolor='black')
plt.title('Volume of Stocks Trading')
plt.show()
```

```python
# Histogram shows distribution of ONE column
plt.hist(data['high'], bins=10)
plt.title('Distribution of High Prices')
plt.show()
```

```python
# Boxplots Compares Several Columns  
plt.boxplot([data['open'], data['high'], data['low'], data['close']])
plt.xticks([1,2,3,4], ['open','high','low','close'])
plt.title('Price Comparison')
plt.show()
```

```python
# Scatter Plots shows relationship between TWO columns
# Its is used for seeing relationships i.e. correlation, trends

plt.scatter(data['high'], data['low'])
plt.xlabel('High')
plt.ylabel('Low')
plt.title('High vs Low Stock Prices')
plt.show()
```

****Correlation** **measures how strongly two variables move together.

**Positive correlation**: both increase together 
**Negative correlation**: one increases while the other decreases
**No correlation**: no consistent relationship

```python
# Correlation Matrix is used to find correlations between numerical features

features = ['open', 'high', 'low', 'close', 'volume']

corr_matrix = data[features].corr()
print(corr_matrix)
```
